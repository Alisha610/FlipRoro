1.) Scrape the details of most viewed videos on YouTube from Wikipedia:
1. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
2. You need to find following details: Rank, Name, Artist, Upload date, Views

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
video_dict = {}
video_dict['Rank']=[]
video_dict['Name']=[]
video_dict['Artist']=[]
video_dict['Upload Date']=[]
video_dict['Views']=[]
```

```python
from selenium.common.exceptions import NoSuchElementException

for i in range(1,31):
    # Extracting Rank from xpath
    try:
        rank = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr["+str(i)+"]/td[1]")
        video_dict['Rank'].append(rank.text)
    except NoSuchElementException:
        video_dict['Rank'].append('-')

    # Extracting Name from xpath
    try:
        name = driver.find_element_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr['+str(i)+']/td[2]/a')
        video_dict['Name'].append(name.text)
    except NoSuchElementException:
        video_dict['Name'].append('-')

    # Extracting Artist from xpath
    try:
        artist = driver.find_element_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr['+str(i)+']/td[3]/a')
        video_dict['Artist'].append(artist.text)
    except NoSuchElementException:
        video_dict['Artist'].append('-')

    # Extracting Upload Date from xpath
    try:
        date = driver.find_element_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr['+str(i)+']/td[5]')
        video_dict['Upload Date'].append(date.text)
    except NoSuchElementException:
        video_dict['Upload Date'].append('-')

    # Extracting Views from xpath
    try:
        views = driver.find_element_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr['+str(i)+']/td[4]')
        video_dict['Views'].append(views.text)
    except NoSuchElementException:
        video_dict['Views'].append('-')
```

```python
video_df = pd.DataFrame.from_dict(video_dict)
video_df
```

```python
#saving data to csv
video_df.to_csv(r'Top Youtube Videos.csv',index=False)
```

5.) Scrape the details of trending repositories on Github.com.
1. Url = https://github.com/
2. You have to find the following details: A) Repository title, B) Repository description, C) Contributors count, D) Language used
3. Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
driver.get("https://github.com/")
time.sleep(3)
```

```python
trending = driver.find_element_by_xpath('//div[@class="dropdown-menu flex-auto rounded-1 color-bg-primary px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4"]/ul[2]/li[3]/a')
url = trending.get_attribute('href')
driver.get(url) 
time.sleep(3)
```

```python
git_dict = {}
git_dict['Repository Title']=[]
git_dict['Repository description']=[]
git_dict['Contributors count']=[]
git_dict['Language used']=[]
```

```python
from selenium.common.exceptions import NoSuchElementException
# Extracting Repository from xpath
repository = driver.find_elements_by_xpath('//article[@class="Box-row"]')
p = len(repository)
for i in range(1,p+1):
    try:
        repository_title = driver.find_element_by_xpath('//div[@class="Box"]/div[2]/article['+str(i)+']/h1[1]/a')
        git_dict['Repository Title'].append(repository_title.text)
    except NoSuchElementException:
        git_dict['Repository Title'].append('-')
        
    try:
        repository_desc = driver.find_element_by_xpath('//div[@class="Box"]/div[2]/article['+str(i)+']/p')
        git_dict['Repository description'].append(repository_desc.text)
    except NoSuchElementException:
        git_dict['Repository description'].append('-')
        
    try:
        lang = driver.find_element_by_xpath('//div[@class="Box"]/div[2]/article['+str(i)+']/div[2]/span[1]/span[2]')
        git_dict['Language used'].append(lang.text)
    except NoSuchElementException:
        git_dict['Language used'].append('-')
        
    try:
        contributors = driver.find_elements_by_xpath('//div[@class="Box"]/div[2]/article['+str(i)+']/div[2]/span[2]/a')
        n = 0
        for j in contributors:
            n += 1
        git_dict['Contributors count'].append(n)
    except NoSuchElementException:
        git_dict['Contributors count'].append('-')
```

```python
git_df = pd.DataFrame.from_dict(git_dict)
git_df
```

```python
#saving data to csv
git_df.to_csv(r'Trending Repositories.csv',index=False)
```
